```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint
```

```python
# Phase portrait system A

def system_A(x,y):
    f_dx = y
    f_dy = -x+y-y**3
    return [f_dx, f_dy]

x_vals = np.linspace(-4, 4, 20)  
y_vals = np.linspace(-6, 6, 20)  

X, Y = np.meshgrid(x_vals,y_vals)
dX_dt = Y
dY_dt = -X + Y - Y**3

theta = np.linspace(0, 2*np.pi, 100)  # 100 points around the circle

# Define the radius of the circle
radius = 1.0

# Calculate the x and y coordinates of the circle
x = radius * np.cos(theta)
y = radius * np.sin(theta)


plt.figure(figsize=(8, 6))
plt.streamplot(X, Y, dX_dt, dY_dt, color='b', linewidth=1, arrowsize=1, density=1.5)
plt.plot(x, y, color='black', linewidth=1.5)
plt.plot([-3,0], [4,5], color='black')
plt.plot([0,3], [5,0], color='black')
plt.plot([3,3], [0,-4], color='black')
plt.plot([3,0], [-4,-5], color='black')
plt.plot([0,-3], [-5,0], color='black')
plt.plot([-3,-3], [0,4], color='black')
plt.xlabel('x')
plt.ylabel('y')
plt.xlim([-4, 4])
plt.ylim([-6, 6])
plt.grid(True)
plt.show()
```

```python
# Phase portrait system B

def system_A(x,y):
    f_dx = y
    f_dy = -x-y-y**3
    return [f_dx, f_dy]

x_vals = np.linspace(-3, 3, 20)  
y_vals = np.linspace(-3, 3, 20)  

X, Y = np.meshgrid(x_vals,y_vals)
dX_dt = Y
dY_dt = -X - Y - Y**3

plt.figure(figsize=(8, 6))
plt.streamplot(X, Y, dX_dt, dY_dt, color='b', linewidth=1, arrowsize=1, density=1.5)
plt.xlabel('x')
plt.ylabel('y')
plt.xlim([-3, 3])
plt.ylim([-3, 3])
plt.grid(True)
plt.show()
```
