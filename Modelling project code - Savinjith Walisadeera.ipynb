```python
import numpy as np
import matplotlib.pyplot as plt

# K2 > r1/alpha1 , K1 > r2/alpha2

r1 = 5
r2 = 10
K1 = 10
K2 = 10
alpha1 = 1
alpha2 = 2
xlim = -10
ylim = 10

def f(Y, t):
    
    x1, x2 = Y
    return [r1 * x1 *(1-x1/K1) - alpha1*x1*x2, r2*x2*(1-x2/K2)-alpha2*x2*x1]

x1 = np.linspace(-2.0, 12.0, 20)
x2 = np.linspace(-2.0, 12.0, 20)

Y1, Y2 = np.meshgrid(x1, x2)

t = 0

u, v = np.zeros(Y1.shape), np.zeros(Y2.shape)

NI, NJ = Y1.shape

for i in range(NI):
    for j in range(NJ):
        x = Y1[i, j]
        y = Y2[i, j]
        yprime = f([x, y], t)
        u[i,j] = yprime[0]
        v[i,j] = yprime[1]
     
     

Q = plt.streamplot(Y1, Y2, u, v, color='r')
def zer(x):
    return 0 * x

def Line1(x):
    return r1/alpha1 - r1*x/(K1*alpha1)

def Line2(x):
    return K2 - x * K2 * alpha2 / r2

plt.axvline(x = 0)
plt.axhline(y = 0)
plt.plot(x2, Line1(x2), color = 'blue', label = '$l_1$')
plt.plot(x2, Line2(x2), color = 'green', label = '$l_2$')

plt.xlabel('$x_1$')
plt.ylabel('$x_2$')
plt.xlim([-2, 10])
plt.ylim([-2, 12])
plt.legend(loc='upper right')
```
